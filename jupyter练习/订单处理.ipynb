```python
import pandas as pd
```

```python
#由于用read_csv()方法，文件名有中文时会报错，需要指定engine为Python
path = r"C:\Users\x\Desktop\测试\订单.csv"
df = pd.read_csv(path,engine='python')
```

```python
df.columns
```

```python
#字段名部分尾部包含空格，导致选取字段时出错，去除字段名首尾的空格
df.columns = df.columns.str.strip()
```

```python
dfs = df[['订单编号','买家会员名','总金额','订单状态','收货地址','订单付款时间','宝贝标题']]
dfs.set_index('订单编号',inplace=True)
dfs = dfs.dropna()
```
