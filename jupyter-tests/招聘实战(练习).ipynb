```python
import pandas as pd
import numpy as np

df = pd.read_csv('DataAnalyst.csv', encoding = 'gbk')
df
```

```python
len(df.positionId.unique())
```

```python
df_duplicates = df.drop_duplicates(subset='positionId',keep='first')
```

```python
def word_cut(word, method):
    position = word.find('-')
    if position != -1:
        bottomSalary = word[:position-1]
        topSalary = word[position+1:-1]
    else:
        bottomSalary = word[:word.upper().find('K')]
        topSalary = bottomSalary
    if method == 'top':
        return topSalary
    else:
        return bottomSalary
```

```python
df_duplicates['bottomSalary'] = df_duplicates.salary.apply(word_cut,method='bottom')
df_duplicates['topSalary'] = df_duplicates.salary.apply(word_cut,method='top')
```

```python
df_duplicates[['companyFullName','bottomSalary','topSalary']]
```

```python
df_duplicates.bottomSalary = df_duplicates.bottomSalary.astype('int')
df_duplicates.topSalary = df_duplicates.topSalary.astype('int')

df_duplicates['avgSalary'] = df_duplicates.apply(lambda x: (x.bottomSalary+x.topSalary)/2,axis=1)
```
